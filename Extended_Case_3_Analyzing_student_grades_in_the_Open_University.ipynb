```python
from pandasql import sqldf
import pandas as pd
dataframes = pd.read_excel('https://workspace-amz-da6.c1-learning.com/extended.tableau_dashboards_fellow/data/OULA.xlsx', sheet_name = None)
student_info = dataframes['student_info']
modules = dataframes['modules']
assessments = dataframes ['assessments']
student_assessment = dataframes ['student_assessment']
```

```python
ex1 = sqldf('SELECT s.id_assessment, s.id_student, s.score, a.code_module, a.weight FROM student_assessment s LEFT JOIN assessments a ON s.id_assessment=a.id_assessment')
ex1
```

```python
ex2 = sqldf('SELECT s.id_assessment, s.id_student,s.score, a.code_module, a.weight FROM student_assessment s LEFT JOIN assessments a ON s.id_assessment= a.id_assessment')
ex2
```

```python
ex3 = sqldf('SELECT SUM(weight) sums, code_module FROM assessments GROUP BY code_module')
ex3
```

```python
ex4 = sqldf('SELECT s.id_assessment, s.id_student,s.score,s.code_module, s.weight, e.sums as sums FROM ex1 s LEFT JOIN ex3 e ON s.code_module = e.code_module')
ex4
```

```python
ex5 = sqldf('SELECT s.id_assessment, s.id_student,s.score, s.code_module, s.weight, s.sums, s.score*s.weight as scoreweight, (s.score*s.weight)/s.sums as scoreweightbysum FROM ex4 s')
ex5
```

```python
query = '''SELECT s.id_student, 
         SUM(CASE WHEN s.code_module = "AAA" then scoreweightbysum end) as AAA,
         SUM(CASE WHEN s.code_module = "BBB" then scoreweightbysum end) as BBB,
         SUM(CASE WHEN s.code_module = "CCC" then scoreweightbysum end) as CCC,
         SUM(CASE WHEN s.code_module = "DDD" then scoreweightbysum end) as DDD,
         SUM(CASE WHEN s.code_module = "EEE" then scoreweightbysum end) as EEE,
         SUM(CASE WHEN s.code_module = "GGG" then scoreweightbysum end) as GGG from ex5 s group by s.id_student'''
ex6 = sqldf(query)
ex6
```

```python
ex7 = sqldf('SELECT s.code_module, s.id_student, SUM(e.scoreweightbysum) FROM student_info s LEFT JOIN ex5 e ON s.id_student = e.id_student GROUP BY s.code_module, s.id_student')
ex7
```
